# **Support Vector Boundary**




Información sobre base de datos isotrópica [aquí](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html#sklearn.datasets.make_blobs) \\
Información sobre support vector classifier [aquí](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC)

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_blobs # Libreria para crear datos sintéticos
```

```
# we create 40 separable points
X, y = make_blobs(n_samples=100, centers=2, random_state=6)

_, ax = plt.subplots (figsize=(5,4))
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap="Paired")
print (X.shape, y.shape)
```

```
svm = SVC(kernel='linear')

svm.fit( X , y )
```

```
slide = 0.1

h_min, h_max = X[:, 0].min()-slide, X[:, 0].max()+slide
v_min, v_max = X[:, 1].min()-slide, X[:, 1].max()+slide

# Crear el grid de las fronteras de decision
h_grid, v_grid = np.meshgrid(np.linspace( h_min , h_max , 300), np.linspace( v_min , v_max , 300))

# Predecir los valores del grid
predictions = svm.predict(np.c_[h_grid.ravel(), v_grid.ravel()])
print (predictions.shape)

# Encajar las predicciones al tamaño del grid
predictions = predictions.reshape(h_grid.shape)
print (predictions.shape)
```

```
_, ax = plt.subplots(figsize=(5,4))

ax.pcolormesh (h_grid, v_grid, predictions, cmap="Paired")
ax.contour(h_grid, v_grid, predictions, linestyles="--", levels=[-1,0,1], colors='k')

ax.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap="Paired")

ax.set_xticks([]), ax.set_yticks([])
```

```
predictions = svm.decision_function(np.c_[h_grid.ravel(), v_grid.ravel()])
print(predictions.shape)
predictions = predictions.reshape(h_grid.shape)
print(predictions.shape)
```

```
_, ax = plt.subplots (figsize=(8,5))
ax.scatter( X[:, 0] , X[:, 1] , c = y , s = 30, cmap = "Paired")

# plot decision boundary and margins
ax.contour( h_grid , v_grid , predictions , colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])

# plot support vectors
ax.scatter(svm.support_vectors_[:, 0], svm.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')

# ax.set_xlim(x_grid.min()-.1, x_grid.max()+.1)
# ax.set_ylim(y_grid.min()-.1, y_grid.max()+.1)

ax.set_xticks(()), ax.set_yticks(())
```
